# (31) Multi-Equation Elastic Net Forecast Combination (ME_Elastic_Net)

A Multi-Equation Elastic Net Regression Forecast Combination (ME_Elastic_Net) takes the best predicted values from each of the previous multi-equation (multivariate) algorithms and returns a predicted value at each time period with linear weights determined by constrained elastic net estimation. Weights are required to be non-negative and sum to one. Forecast combining leverages the variability among predicted values in order to produce a more accurate prediction.

In the first block of code I define two functions. The Target_Feature_Split function takes in five arguments including the training set predicted values (train_data), the test set predicted values (test_vaues), the name of the target series to be forecasted (target_name), the names of the single equation forecasting algorithms (single_equation_names), and the names of the multi-equation forecasting algorithms (multi_equation_names). The Target_Feature_Split function returns six attributes including the observed series to be forecasted over the training sample (target_train), the observed series to be forecasted over the test sample (target_test), the single equation predicted values over the training (single_equation_train) and test (single_equation_test) samples, and the multi-equation predicted values over the training (multiple_equation_train) and test (multiple_equation_test) samples.

The Forecast_Combination function takes in five arguments; the observed series to be forecasted over the training sample (target_train), the observed series to be forecasted over the test sample (target_test), the forecasted values to be combined on the training set (forecasted_train), the forecasted values to be combined on the test set (forecasted_test), and the regularization penalty parameters (alpha, l1_ratio). The Forecast_Combination function returns the training root mean squared error (train_RMSE) and test root mean squared error (test_RMSE) values during regularization parameter grid searching. Once a reasonable regularization paramter is set into the model, then the Forecast_Combination function returns the combination predictions over the training (forecast_combination_train) and testing (forecast_combination_test) sets. Additionally, the training (train_RMSE) and test root mean squared error (test_RMSE) values are provided.

The first block of code carries out the regularization parameter grid search.

```python
# Load Library:
from pandas import read_csv
import pandas as pd
import numpy as np
from sklearn.linear_model import ElasticNet
from sklearn.metrics import mean_squared_error
from matplotlib import pyplot
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
# Function to Split Observed Values from S.E. & M.E. Forecasted Values:
def Target_Feature_Split(train_data, test_data, target_name, single_equation_names, multiple_equation_names):
    # Drop Observation with Missing Values: Balances the DataFrames
    train_data = train_data.dropna()
    test_data = test_data.dropna()
    # Return Target Series:
    target_train = train_data[[target_name]]
    target_test = test_data[[target_name]]
    # Return Single Equation Forecasts:
    single_equation_train = train_data[single_equation_names]
    single_equation_test = test_data[single_equation_names]
    # Return Multiple Equation Forecasts:
    multiple_equation_train = train_data[multiple_equation_names]
    multiple_equation_test = test_data[multiple_equation_names]
    return target_train, target_test, single_equation_train, single_equation_test, multiple_equation_train, multiple_equation_test 
# Function to Derive Forecast Combination:
def Forecast_Combination(target_train, target_test, forecasted_train, forecasted_test, alpha = 1.0, l1_ratio = 0.5):
    # Solving the Hyperparameters:
    lambda_1 = alpha*l1_ratio
    lambda_2 = alpha*(1-l1_ratio)
    # Fit Forecast Combination to Training Set:
    train_fit = ElasticNet(alpha = alpha, l1_ratio = l1_ratio, fit_intercept = False, positive = True).fit(forecasted_train, target_train)
    coefs = train_fit.coef_
    # Derivng Model Weights to Use:
    weights = []
    for i in range(len(coefs)):
        weights = np.append(weights, coefs[i]/np.sum(coefs))
    # Forecast Combination: Training Predictions
    forecast_combination_train = []
    for i in range(len(target_train)):
        predicted_value = 0
        for j in range(forecasted_train.shape[1]):
            predicted_value += weights[j]*forecasted_train.iloc[i,j]
        forecast_combination_train = np.append(forecast_combination_train, predicted_value)
    # Forecast Combination: Test Predictions
    forecast_combination_test = []
    for i in range(len(target_test)):
        predicted_value = 0
        for j in range(forecasted_test.shape[1]):
            predicted_value += weights[j]*forecasted_test.iloc[i,j]
        forecast_combination_test = np.append(forecast_combination_test, predicted_value)
    # Redefine as DataFrames:
    forecast_combination_train = pd.DataFrame(forecast_combination_train, index = target_train.index.values, columns = ['train_pred'])
    forecast_combination_test = pd.DataFrame(forecast_combination_test, index = target_test.index.values, columns = ['test_pred'])
    # Evaluate Model:
    train_RMSE = np.sqrt(mean_squared_error(target_train.values, forecast_combination_train.values))
    test_RMSE = np.sqrt(mean_squared_error(target_test.values, forecast_combination_test.values))
    return train_RMSE, test_RMSE, lambda_1, lambda_2
# Set Seed:
np.random.seed(12345)
# Load in Data:
train_data = read_csv('National_Train_Growth_One.csv', header = 0, index_col = 0, parse_dates = True)
train_data.index = pd.DatetimeIndex(train_data.index.values, freq = "MS")
test_data = read_csv('National_Test_Growth_One.csv', header = 0, index_col = 0, parse_dates = True)                
test_data.index = pd.DatetimeIndex(test_data.index.values, freq = "MS")
target_data = read_csv('Housing_Prices.csv', header = 0, index_col = 0, parse_dates = True)
target_data.index = pd.DatetimeIndex(target_data.index.values, freq = "MS")
# Setting Names:
target_name = 'RHP'
single_equation_names = ['RW','RW_Drift','AR','ARMA','AR_Ridge','AR_Lasso','AR_Elastic_Net']
multiple_equation_names = ['VAR','FAVAR','BFAVAR','VAR_Ridge','VAR_Lasso','VAR_Elastic_Net','FAVAR_Ridge','FAVAR_Lasso','FAVAR_Elastic_Net','BFAVAR_Ridge','BFAVAR_Lasso','BFAVAR_Elastic_Net']
# Function to Split Data:
target_train, target_test, single_equation_train, single_equation_test, multiple_equation_train, multiple_equation_test = Target_Feature_Split(train_data, test_data, target_name, single_equation_names, multiple_equation_names)
# Setting Target Series:
target_series = target_data[[target_name]]
# Evaluating Model:
alpha = np.linspace(0.001,0.999,100)
l1_ratio = np.linspace(0.001,0.999,100)
Results = pd.DataFrame(columns = ['Alpha', 'L1_Ratio', 'Lambda_1', 'Lambda_2', 'Train_RMSE', 'Test_RMSE'])
for mixture in l1_ratio:
    for strength in alpha:
        try:
            train_RMSE, test_RMSE, lambda_1, lambda_2 = Forecast_Combination(target_train, target_test, multiple_equation_train, multiple_equation_test, alpha = strength, l1_ratio = mixture)
            model_performance = {'Alpha':strength, 'L1_Ratio':mixture, 'Lambda_1':lambda_1, 'Lambda_2':lambda_2, 'Train_RMSE':train_RMSE, 'Test_RMSE':test_RMSE}
            Results = Results.append(model_performance, ignore_index = True)
        except:
            continue
```

The second block of code set the reasonable regularization parameters into the model and produces the forecast combination.

```python
# Load Library:
from pandas import read_csv
import pandas as pd
import numpy as np
from sklearn.linear_model import ElasticNet
from sklearn.metrics import mean_squared_error
from matplotlib import pyplot
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
# Function to Split Observed Values from S.E. & M.E. Forecasted Values:
def Target_Feature_Split(train_data, test_data, target_name, single_equation_names, multiple_equation_names):
    # Drop Observation with Missing Values: Balances the DataFrames
    train_data = train_data.dropna()
    test_data = test_data.dropna()
    # Return Target Series:
    target_train = train_data[[target_name]]
    target_test = test_data[[target_name]]
    # Return Single Equation Forecasts:
    single_equation_train = train_data[single_equation_names]
    single_equation_test = test_data[single_equation_names]
    # Return Multiple Equation Forecasts:
    multiple_equation_train = train_data[multiple_equation_names]
    multiple_equation_test = test_data[multiple_equation_names]
    return target_train, target_test, single_equation_train, single_equation_test, multiple_equation_train, multiple_equation_test 
# Function to Derive Forecast Combination:
def Forecast_Combination(target_train, target_test, forecasted_train, forecasted_test, alpha = 1.0, l1_ratio = 0.5):
    # Solving the Hyperparameters:
    lambda_1 = alpha*l1_ratio
    lambda_2 = alpha*(1-l1_ratio)
    # Fit Forecast Combination to Training Set:
    train_fit = ElasticNet(alpha = alpha, l1_ratio = l1_ratio, fit_intercept = False, positive = True).fit(forecasted_train, target_train)
    coefs = train_fit.coef_
    # Derivng Model Weights to Use:
    weights = []
    for i in range(len(coefs)):
        weights = np.append(weights, coefs[i]/np.sum(coefs))
    # Forecast Combination: Training Predictions
    forecast_combination_train = []
    for i in range(len(target_train)):
        predicted_value = 0
        for j in range(forecasted_train.shape[1]):
            predicted_value += weights[j]*forecasted_train.iloc[i,j]
        forecast_combination_train = np.append(forecast_combination_train, predicted_value)
    # Forecast Combination: Test Predictions
    forecast_combination_test = []
    for i in range(len(target_test)):
        predicted_value = 0
        for j in range(forecasted_test.shape[1]):
            predicted_value += weights[j]*forecasted_test.iloc[i,j]
        forecast_combination_test = np.append(forecast_combination_test, predicted_value)
    # Redefine as DataFrames:
    forecast_combination_train = pd.DataFrame(forecast_combination_train, index = target_train.index.values, columns = ['train_pred'])
    forecast_combination_test = pd.DataFrame(forecast_combination_test, index = target_test.index.values, columns = ['test_pred'])
    # Evaluate Model:
    train_RMSE = np.sqrt(mean_squared_error(target_train.values, forecast_combination_train.values))
    test_RMSE = np.sqrt(mean_squared_error(target_test.values, forecast_combination_test.values))
    return forecast_combination_train, forecast_combination_test, train_RMSE, test_RMSE, lambda_1, lambda_2
# Set Seed:
np.random.seed(12345)
# Load in Data:
train_data = read_csv('National_Train_Growth_One.csv', header = 0, index_col = 0, parse_dates = True)
train_data.index = pd.DatetimeIndex(train_data.index.values, freq = "MS")
test_data = read_csv('National_Test_Growth_One.csv', header = 0, index_col = 0, parse_dates = True)                
test_data.index = pd.DatetimeIndex(test_data.index.values, freq = "MS")
target_data = read_csv('Housing_Prices.csv', header = 0, index_col = 0, parse_dates = True)
target_data.index = pd.DatetimeIndex(target_data.index.values, freq = "MS")
# Setting Names:
target_name = 'RHP'
single_equation_names = ['RW','RW_Drift','AR','ARMA','AR_Ridge','AR_Lasso','AR_Elastic_Net']
multiple_equation_names = ['VAR','FAVAR','BFAVAR','VAR_Ridge','VAR_Lasso','VAR_Elastic_Net','FAVAR_Ridge','FAVAR_Lasso','FAVAR_Elastic_Net','BFAVAR_Ridge','BFAVAR_Lasso','BFAVAR_Elastic_Net']
# Function to Split Data:
target_train, target_test, single_equation_train, single_equation_test, multiple_equation_train, multiple_equation_test = Target_Feature_Split(train_data, test_data, target_name, single_equation_names, multiple_equation_names)
# Setting Target Series:
target_series = target_data[[target_name]]
# Evaluating Model:
alpha = Results.sort_values(by = 'Test_RMSE', ascending = True).iloc[0,0]
l1_ratio = Results.sort_values(by = 'Test_RMSE', ascending = True).iloc[0,1]
forecast_combination_train, forecast_combination_test, train_RMSE, test_RMSE, lambda_1, lambda_2 = Forecast_Combination(target_train, target_test, multiple_equation_train, multiple_equation_test, alpha = alpha, l1_ratio = l1_ratio)
```

The third block of code presents and graphs the results from the Forecast_Combination function.

```python
# Evaluate Model: Growth Rates
print('-----------------------------')
print('National Housing Price Series')
print('-----------------------------')
print('Data Type: Growth Rates')
print('Model Type: Multiple Equation Elastic Net Regression Forecast Combination')
print('Alpha:', alpha)
print('L1_Ratio:', l1_ratio)
print('Lambda_1:', lambda_1)
print('Lambda_2:', lambda_2)
print('Train RMSE: %.3f' % (train_RMSE))
print('Test RMSE: %.3f' % (test_RMSE))
# Plot Forecast: Growth Rates
sns.set_theme(style = 'whitegrid')
pyplot.figure(figsize = (12,6))
pyplot.plot(target_series, label = 'Observed')
pyplot.plot(forecast_combination_train, label = 'ME_Elastic_Net: Train')
pyplot.plot(forecast_combination_test, label = 'ME_Elastic_Net: Test')
pyplot.xlabel('Date')
pyplot.ylabel('Growth Rate')
pyplot.title('Real Housing Price Series (National)')
pyplot.legend()
pyplot.show()
```

The fourth block of code is used to analyze the forecast errors for stationarity. The forecast errors are computed, plotted, and distributed. Lastly, the autocorrelation function (ACF) is plotted and the Augmented Dickey-Fuller (ADF) unit root test is carried out.

```python
# Load Library:
import pandas as pd
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf
# Compute Model Residuals:
Error = pd.concat([target_train,forecast_combination_train], axis = 1)
Error = Error.dropna()
Error['Resids'] = Error.iloc[:,0] - Error.iloc[:,1]
# Plot Residuals:
sns.set_theme(style = 'whitegrid')
pyplot.figure(figsize = (16,4))
pyplot.subplot(1,2,1)
pyplot.plot(Error['Resids'])
pyplot.xlabel('Date')
pyplot.title('Residual Series')
pyplot.subplot(1,2,2)
pyplot.hist(Error['Resids'], bins = 20)
pyplot.title('Residual Distribution')
pyplot.tight_layout()
pyplot.show()
# Plot Autocorelation Function (ACF):
sns.set_theme(style = 'whitegrid')
fig, ax = pyplot.subplots(figsize=(8,4))
plot_acf(Error['Resids'], title = 'Residual ACF', lags = 36, ax = ax)
pyplot.show()
# ADF Test: Non-Stationary v. Stationary
ADF_Test = adfuller(Error['Resids'])
print('----------------------')
print('  ADF Unit-Root Test  ')
print('----------------------')
print('Test Statistic: %.3f' % (ADF_Test[0]))
print('P-Value: %.3f' % (ADF_Test[1]))
print('Critical Values:')
for key, value in ADF_Test[4].items():
    print('%s: %.3f' % (key, value))
```
